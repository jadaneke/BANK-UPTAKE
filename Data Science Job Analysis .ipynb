## Exploratory Data Analysis (EDA)

- Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, usually by employing data visualization methods. 
- It helps determine how best to manipulate data sources to get the answers you need, making it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.
-We have two types of EDA namely Univariate and Multivariate

```python
# Data Analysis libraries
import numpy as np
import pandas as pd
# Visualisation libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

## Univariate Analysis using subplot

```python
fig, axes = plt.subplots(2,4, figsize=(20,10))

# Categorical variables

sns.countplot(x= 'experience_level', data=df, ax=axes[0,0])
sns.countplot(x= 'job_title', data=df, ax=axes[0,1])
sns.countplot(x= 'employment_type', data=df, ax=axes[0,2])
sns.countplot(x= 'employee_residence', data=df, ax=axes[0,3])
sns.countplot(x= 'company_size', data=df, ax=axes[1,0])


# Numerical data

sns.histplot(x= 'work_year', data=df, ax=axes[1,1])
sns.histplot(x= 'salary_in_usd', data=df, ax=axes[1,2])
sns.histplot(x= 'remote_ratio', data=df, ax=axes[1,3])
```

```python
plt.figure(figsize=(9,8))
plt.title('Work Year Analysis')
sns.countplot(x='work_year', data=df)
plt.xlabel('work_year')
plt.ylabel('remote_ratio')
plt.show()
```


##  2. - Work year Analysis: Remote Ratio by Work Year

## Salary Analysis:

## 1 - Salary by work year  plotting visual charts:


```python
salary_in_usd = df['salary'].values
work_year = df['work_year'].values

plt.title('Salary Analysis By Work Year')
plt.bar(x=work_year,height=salary_in_usd)
plt.xlabel('Work Year')
plt.ylabel('Salary in USD')
plt.show()
```

## 2 - Salary by Experience Level

```python
salary_in_usd = df['salary'].values
experience_level = df['experience_level'].values

plt.title('Salary Analysis By Experience Level')
plt.bar(x=experience_level,height=salary_in_usd)
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

## 3 - Salary by Company Size

```python
salary_in_usd = df['salary'].values
company_size = df['company_size'].values

plt.title('Salary Analysis By Company Size')
plt.bar(x=company_size,height=salary_in_usd)
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.show()
```

## 4 - Highest Salaries By Job Title

```python
salary_in_usd = df['salary'].values
job_title = df['job_title'].values

plt.figure(figsize=(20,10))
plt.title('Highest Salaries By Job Title')
plt.bar(x=job_title,height=salary_in_usd)
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')
plt.show()
```

## 5 - Average Salary by Company Location

```python
salary_in_usd = df['salary'].values
company_location = df['company_location'].values

plt.figure(figsize=(20,10))
plt.title('Averege Salary By Company Location')
plt.bar(x=company_location,height=salary_in_usd)
plt.xlabel('Company Location')
plt.ylabel('Average Salary in USD')
plt.show()
```
